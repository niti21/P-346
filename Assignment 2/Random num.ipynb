```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
def randomwalk(n,p,q):
    x = 0
    y = 0
    x1 = 0
    y1 = 0
    a_x1=[]
    a_y1=[]
    r = 0
    s = 0
    
    for i in range(0,n):

        #Random walk in x-axis
        t1 =  lcglib.lcg(1103515245, 12345, 32768,p)
        p=t1
        t1 = 2*(math.pi)*t1
        
        #Random walk in y-axis
        t2 = lcglib.lcg(1103515245, 12345, 32768,q)
        q=t2
        t2 = 2*(math.pi)*t2

        #new coordinates
        x1 = x1 + math.cos(t1)
        y1 = y1 + math.sin(t2)


        a_x1.append(x1)
        a_y1.append(y1)

        #for calculating RMS distance 
        r = (math.cos(t1)**2)+(math.sin(t2)**2)
        s = s + r



    plt.plot(a_x1, a_y1, 'o-')


    d = math.sqrt(((a_x1[n-1]**2)-(x**2))+((a_y1[n-1]**2)-(y**2)))
    print('The net displacement:',d)

    rms = math.sqrt(s/n)
    print('The RMS distance covered:',rms)
```

```python
#defining seeds
p1 = 101
q1 = 21        
randomwalk(300,p1,q1)
#for N=300
```
