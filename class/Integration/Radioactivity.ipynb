```python

import math
import NSlib as l
import matplotlib.pyplot as plt

def DecayABC(Na,Nb,Nc,Ta,Tb,T,dT):
    lA = math.log(2)/Ta
    lB = math.log(2)/Tb
   
    r = 10
    t = 0
   
    NA=[Na]
    NB=[0]
    NC=[0]
    NT=[0]
    i=0
    while i in range(0,int(T/dT)) and Na >=0:
       
        for j in range(0,Na):
            if r <= lA*dT:
                Na = Na - 1
                Nb = Nb + 1
       
            r = l.random(r)
        for j in range(0,Nb):
            if r <= lB*dT:
                Nb = Nb - 1
                Nc = Nc + 1
       
            r = l.random(r)
       
        NA.append(Na)
        NB.append(Nb)
        NC.append(Nc)
        t +=dT
        NT.append(t)
        i=i+1
   
    return NA,NB,NC,NT
  
```

```python
nA=500
tA=20
tB=50
T=500

NA,NB,NC,NT = DecayABC(nA,0,0,tA,tB,T,2)
plt.title("Radioctive series decay ")
plt.plot(NT,NA,label='A')
plt.plot(NT,NB,label='B')
plt.plot(NT,NC,label='C')
# Label for x-axis
plt.xlabel("dN/dT")
# Label for y-axis
plt.ylabel("dN/dT")
plt.legend()
plt.show()
```
