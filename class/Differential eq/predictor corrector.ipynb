```python
import math
import numpy as np
import matplotlib.pyplot as plt

def y_prime(x):
    return(math.exp(x))
   
def perdictor_corrector(x0,y0,N,x1,f):
    h=(x1-x0)/N
    x = x0
    y = y0
    X = [x0]
    Y = [y0]
    while x<x1:
        k1 = h*f(x)
        k2 = h*f(x+h)
        y = y+(k1+k2)/2
        x = x+h
        X.append(x)
        Y.append(y)
    return X,Y


X1 = np.linspace(0,3,100)
Y1 = np.exp(X1)

X,Y = perdictor_corrector(0,1,5,3,y_prime)

plt.plot(X1,Y1, color = 'r',)        
plt.plot(X, Y, color = 'b',)    
plt.show()
```
