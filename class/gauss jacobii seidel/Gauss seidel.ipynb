```python
def arrclose(x,y,tol):
    count=0
    if len(x)==len(y):
        for i in range(0,len(x)):
            if (abs(x[i]-y[i]))/abs(y[i]) < tol:
                count =count+1
            else:
                return False
    if count==len(x):
        return True

def gsm(a,x,b):
    for j in range(0, len(a)):               
          
        d = b[j]
        
        for i in range(0, len(a)):     
            if(j != i):
                
                d -= a[j][i] * x[i]
        
        x[j] = d / a[j][j]
        
    return x

def GaussSeidel(a,x,b,it,tol):
    
    y=[]
    
    for i in range(0,len(x)):
        y.append(x[i])
    
    for i in range(1,it):
        t=0
        for t in range(0,len(x)):
            y[t] = x[t]
    
        x = gsm(a,x,b)
    
        print(i,':',x)
    
        z = arrclose(x,y,tol)
        if z == True:
            print()
            print('result:',x)
            break
        else:
            continue
```
