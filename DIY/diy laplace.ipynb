```python
# Numerical Laplace Equation Solution using Finite Difference Method
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import diylib as l

# Maximum iteration, Dimension and delta
dimx = 10
dimy = 10
it = 500
dt = 1

# Boundary condition
Vt = -20
Vb = 20
Vl = 0
Vr = 0

# Initial guess of interior grid
Vg = 30

print('Potential Gradient for Capacitor as shown in colorbar in Volts')
print()

X,Y,V = l.Laplace(Vt,Vb,Vl,Vr,Vg,dt,dimx,dimy,it)

nt = plt.axes(projection='3d')
nt.contour3D( Y, X, V)
# Label for x-axis
nt.set_xlabel("y")
# Label for y-axis
nt.set_ylabel("x")
# Label for z-axis
nt.set_zlabel("Potential")
nt.set_title('3D contour plot for potential of a Parallel plate capacitor')
plt.show()
```
